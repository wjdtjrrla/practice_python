```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
```

```python
url = 'https://comp.wisereport.co.kr/company/c1010001.aspx?cmp_cd='
cod = '012450'
```

```python
res = requests.get(url+cod)
```

```python
res.text
```

```python
# res 데이터에서 text로 추출한 뒤 파싱
soup = bs(res.text, 'html.parser')
```

```python
len(
    soup.find_all(
        'table',
        attrs = {
            'class':'gHead01 all-width'
            }
        )
)
```

```python
# table 태그 중 class가 gHead01 all-width인 태그 중 마지막 데이터만 사용
table_data = \
    soup.find_all(
        'table',
        attrs = {
            'class' : 'gHead01 all-width'
        }
    )[-1]
```

```python
table_data
```

```python
# table_data를 문자열로 변환 데이터프레임으로 만든다.
df = pd.read_html(
    str(table_data)
)[-1]
```

```python
url = 'https://comp.wisereport.co.kr/company/c1010001.aspx?cmp_cd='
cod = '012450'
```

```python
driver = webdriver.Chrome()
driver.get(url+cod)
```

```python
soup = bs(driver.page_source,'html.parser')
```

```python
len(
    soup.find_all(
        'table',
        attrs = {
            'class':'gHead01 all-width'
            }
        )
)
```

```python
table_data = soup.find_all(
        'table',
        attrs = {
            'class':'gHead01 all-width'
            }
        )[-1]
```

```python
df = pd.read_html(
    str(table_data)
)[0]
```

```python
df
```
