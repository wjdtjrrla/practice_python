```python
import pandas as pd
```

### 연습문제
1. csv폴더 안에 있는 'Sales Records.csv' 파일 로드 (pd.read_csv())
2. 'Order ID', 'Total Revenue', 'Total Cost','Total Profit','Order Priority' 컬럼 제거 ( drop() )
3. 컬럼의 이름 변경 : ['대륙','국가','상품종류','판매채널','주문일자','발송일자','판매개수','단가','원가']
4. 대륙, 국가를 기준으로 오름차순 sort_values()
5. 파생변수 생성
    - 총단가 -> 판매개수 * 단가
    - 총원가 -> 판매개수 * 원가
    - 총이윤 -> 총단가 - 총원가

```python
sales = pd.read_csv('../csv/Sales Records.csv')
```

```python
sales.info()
#해당 코드로 해당 데이터에는 결측값이 없음을 확인.
#총데이터개수 10000 : 각 데이터에 non-null 1000000
```

```python
#데이터 확인인
sales.head()
```

```python
#특정 컬럼 삭제
df = sales.drop(
    ['Order ID', 'Total Revenue', 'Total Cost','Total Profit','Order Priority'],
    axis = 1
)
```

```python
#컬럼의 이름 변경
df.columns = [
    '대륙',
    '국가',
    '상품종류',
    '판매채널',
    '주문일자',
    '발송일자',
    '판매개수',
    '단가',
    '원가'
    ]
```

```python
df.head(1)
```

```python
df.rename(
    columns = {
        '판매개수' : '개수'
    }
)
```

```python
#대륙, 국가를 기준으로 오름차순
df.sort_values(
    ['대륙','국가'],
    inplace = True
)
```

```python
#인덱스 재정렬
df.reset_index(drop=True,inplace= True)
```

```python
df.head()
```

```python
#파생변수 생성
df['총단가'] = df['판매개수'] * df['단가']
df['총원가'] = df['판매개수'] * df['원가']
df['총이윤'] = df['총단가'] - df['총원가']
```

```python
df.head()
```

```python
#총 이윤의 합계가 가장 높은 국가는?
#먼저 국가를 기준으로 그룹을 만들어보자 !

#국가의 개수
print(len(df['국가'].unique()))

group_data = df[['국가','총이윤']].groupby('국가').sum()
```

```python
#총이윤이 가장 높은 10개의 국가
group_data.sort_values('총이윤',ascending=False).head(10)
```

```python
#해당하는 데이터에서 South Korea의 순위는?

gr = group_data.sort_values('총이윤',ascending=False)
gr.loc[ ['South Korea'] ]
```

```python
rank_n=0
for i in gr.index:
    rank_n += 1
    if i == 'South Korea':
        break
    
print(rank_n)
```

```python
list(gr.index).index('South Korea') + 1
```

```python
#group_data에서 인덱스를 초기화하고 기존의 인덱스를 컬럼으로 유지
gr1 = gr.reset_index()
gr1.loc[gr1['국가'] == 'South Korea']
```

```python
#rank() 함수 이용
gr2 = gr['총이윤'].rank(ascending=0)
gr2
```

```python
gr2.loc['South Korea']
```
