```python
# 환율 정보 csv 저장 자동화 풀 코드

import os

file_name = '환율.csv'

header_bool  = not os.path.exists(file_name)

df = pd.read_html(
    'https://finance.naver.com/marketindex/exchangeList.naver',encoding='cp949'
    )[0]

df.to_csv(
    file_name,
    mode='a',
    index =False,
    header=header_bool
)
```

```python
import database
```

```python
# SQL에 데이터 대입
mydb = database.MyDB()
```

```python
insert_query = '''
    insert into `exchange`(
        `통화명`, `매매기준율`, `현찰_사실때`, `현찰_파실때`, `송금_보내실때`, `송금_받으실때`,`미화환산율` 
    )
    values(%s,%s,%s,%s,%s,%s,%s)
'''
mydb.execute_query(insert_query,*df.loc[0].values,inplace=True)
```

```python
df.fillna(0,inplace=True)

# 데이터의 길이만큼 반복 실행하는 반복문을 생성
for i in range(len(df)):
    mydb.execute_query(insert_query,*df.loc[i].values,inplace=True)
```

```python
from sqlalchemy import create_engine
```

```python
host = '127.0.0.1'
port = 3306
user = 'root'
pwd = '1234'
db = 'ubion'
```

```python
# Sqlalchemy에 엔진 생성
engine = create_engine(f'mysql+pymysql://{user}:{pwd}@{host}:{port}/{db}')
```

```python
df.to_sql(
    name = 'exchange',      # 테이블의 이름 ( 혹시 생성해둔 테이블 이름을 쓰지 않더라도, 새로운 테이블 만들어냄 )
    con = engine,           # 데이터베이스 정보 ( sqlalchemy engine )
    if_exists = 'append',   # 테이블이 존재한다면 추가한다.
    index = False
)
```
