#### 일일확진자, 일일사망자 컬럼을 추가하는 함수
- 매개변수 2개
    - 첫 번째 인자 : 데이터 프레임
    - 두 번째 인자 : 차순정렬을 할 타켓 컬럼
- 특정 컬럼을 제거
    - 'Unnamed: 0','seq' 컬럼을 제거
- 컬럼의 이름을 변경
    - '등록일시', '총사망자','총확진자','기준일','기준시간','수정일시','누적의심자','누적확진율'
- 특정 컬럼의 데이터를 기준으로 오름차순 정렬
- 인덱스를 초기화하고 기존의 인덱스를 제거
- 일일확진자 컬럼을 추가
    - 총확진자의 차이를 가지고 데이터 생성
- 일일사망자 컬럼을 추가
    - 총사망자의 차이를 가지고 데이터 생성
- 최종 결과 데이터프레임을 되돌려준다.

```python
def add_col(_data, _col):
    
    data = _data.copy()
    col = _col
    
    data.drop(
        ["Unnamed: 0",'seq'],
        axis = 1,
        inplace=True
    )
    
    data.columns = [
    '등록일시',
    '총사망자',
    '총확진자',
    '기준일',
    '기준시간',
    '수정일시',
    '누적의심자',
    '누적확진율'
    ]
    
    data.sort_values(
        [col],
        inplace = True
    )
    
    data.reset_index(drop=True,inplace=True)
    
    data['일일확진자'] = \
        (data['총확진자']-data['총확진자'].shift()).fillna(0)
    data['일일사망자'] = \
        data['총사망자'].diff().fillna(0)
    
    return data
    
```
