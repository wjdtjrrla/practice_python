- replace() 함수 생성
    - 1차원 데이터(스리즈)에서 replace()
        - 매개변수(_list,a,b) 3개
        - 1차원 데이터의 원소들을 a와 비교해서 같다면 해당 위치의 데이터를 b로 변경
    - 문자열에서 replace()
        - 매개변수(_list,a,b) 3개
        - 각 데이터의 문자 하나씩 a와 비교해서 같다면 해당 위치의 데이터를 b로 변경

```python
_list_data = ['a','a','ab','c']

def s_replace(_list,_a,_b):
    #_list입력되는 값들 중 a 와 같은 데이터가 존재한다면 변경
    for i in range(len(_list)):
        if _list[i] == _a:
            _list[i] = _b
    return _list
```

```python
s_replace(_list_data,'a','change')
```

```python
#문자열에서 사용되는 replace
#매개변수 3개
#결과 result값으로 ""문자열 생성
#_str의 문자 하나씩 추출해 a와 비교
    #같으면 b를 result에
    #같지 않으면 원래를 result에
#결과 리턴

def text_replace(_str,_a,_b):
    result = ''
    
    for i in _str:
        if i == _a:
            result += _b
        else:
            result += i
    return result

```

```python
def c_replace(_list,_a,_b):
    a=str(_a)
    b=str(_b)
    result = ''
    
    for i in range(len(_list)):
        for j in range(len(_list[i])):
            if _list[i][j] == a:
                result += b
            else:
                result += _list[i][j]
        _list[i] = result
        result = ''
    return _list
    
```

- 상품명에 따라 결측치를 채워준다.
    1. 상품A인 데이터 중 item_price가 결측치가 아닌 데이터들의 item_price의 평균 변수에 저장
    2. 상품A인 데이터 중 item_price가 결측치인 데이터에 평균을 대입
    3. 위의 작업을 상품 A부터 Z까지 반복

```python
#item_name이 상품A인 경우
flag = df2['item_name']=='상품A'

#item_price가 결측치인 경우
flag_null = df2['item_price'].isna()

#item_name이 상품A이며 결측값이 없는 경우
# df2.loc[flag & ~flag_null].describe().loc['mean','item_price']
mean_data = \
    df2.loc[flag & ~flag_null,'item_price'].mean()
```

```python
df2.loc[flag & flag_null].fillna(mean_data)
```

```python
df2.loc[flag&flag_null,'item_price'] = mean_data
```

```python
df2.isna().sum()
```

```python
#상품A~상품Z까지 item_name의 유니크 값 반복 작업
item_list = df2['item_name'].unique()

#item_price가 결측치인 경우
flag_null = df2['item_price'].isna()
    
#반복문 생성
for i in item_list:
    #item_name이 상품A인 경우
    flag = df2['item_name'] == i

    #1번작업(item_price평균)
    #df.loc[flag&flag_null].describe().loc['mean', 'item_price']
    mean_data = \
        df2.loc[flag & ~flag_null, 'item_price'].mean()
    df2.loc[flag & flag_null] =\
        df2.loc[flag & flag_null].fillna(mean_data)
```
