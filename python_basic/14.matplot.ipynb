```python
#데이터 로드
drinks = pd.read_csv('../csv/drinks.csv')
```

```python
drinks
# 세 종류의 상관간계가 어떨까
#   
```

```python
x = drinks['beer_servings']
y = drinks['wine_servings']

plt.scatter(x,y)
plt.show()
```

```python
#데이터에서 상관계수 출력
#corr()
drinks_corr = drinks.iloc[:,1:5].corr(method='pearson')
drinks_corr
```

```python
cols_view = ['beer','spirit','wine','alcohol']

hm = sns.heatmap(
    drinks_corr.values,
    cbar = True, #사이드바 유무
    annot =True, #히트맵에 숫자를 표시할건지
    square = True,#히트맵을 정사각형 생성 유무
    fmt = '.3f', # 소수점 3자리까지 표시
    annot_kws={'size' : 20}, #숫자의 크기
    xticklabels=cols_view,
    yticklabels=cols_view,
    cmap = "Greens"
)
```

```python
#데이터의 분포도 확인
sns.pairplot(
    drinks.loc[:,'beer_servings' : 'total_litres_of_pure_alcohol'],
    height = 1.5,
    corner=True
)
```
