```python
# 매개변수가 존재하는 함수 생성
def func_3(_x,_y):
    # _x, _y가 매개변수 -> 지역 변수의 종류 중 하나
    # 함수 내부에서는 사용 가능, 외부에서는 불가능 
    result = _x+_y
    # result도 지역 변수
    # return 함수에서 생성된 데이터를 외부에서 사용 ㄱ능하게 되돌려주는 부분
    return result
# result는 함수 내부에서 생성된 변수이기 때문에 호출 불가능
```

# 클래스
- 변수 + 함수의 모음
- 변수들을 class 끼리 공유하는 변수가 존재, 독립적인 변수가 존재
- class 내부의 변수 내부의 값은 수정하고 출력하기 위한 객체

```python
class Ramen:
    cate = {
        '면,라면스프':'라면',
        '면,짜장스프':'짜장면',
        '면,라면스프,계란':'계란탁',
        '면,라면스프,치즈':'치즈라면',
        '면,짜장스프,치즈,계란':'짜계치'
    }
    # 생성자 함수 : class가 생성될 때 최초로 한 번 실행하는 함수
    # class 내부에서 사용할 변수의 데이터를 대입하는 과정
    
    # self란? ㄴㄴ
    def __init__(self,a,b):
        self.a=a
        self.b=b
        
    def cooking(self, *option):
        # 조합 a,b
        # 면, 라면스프/짜장스프
        # option -> tuple
        res = self.a+','+self.b
        for i in option:
            res+=','+i
        return Ramen.cate[res]
```

```python
class1=Ramen('면','라면스프')
```

```python
class1.cooking('계란')
```

        - 1차원 데이터의 형태

    - DataFrame
        - 2차원 데이터

        - 1차원 데이터에서 value의 값들을 하나씩 추출하여 특정 함수에 대입한 뒤 결과를 Series변경

            - 일반 함수 사용 가능
            - lambda함수 사용 가능
    - shift(n)
        - 인덱스를 이동하는 함수

    - diff(n)
        - 인덱스를 n만큼 이동한 데이터와의 값의 차이를 출력하는 함수

    - isin()
        - 1차원데이터에 포함되어있는지 여부를 판단하는 함수
    - fillna()
        - 결측치는 다른 값을 대체해주는 함수
        - 인자에 단일 데이터를 입력하면 해당 데이터로 결측치가 대체
        - method 매개변수를 이용
            - ffill : 전 인덱스의 데이터로 결측치 대체 (forward)
            - bfill : 후 인덱스의 데이터로 결측치 대체 (back)
    - groupby()
        - 특정 컬럼의 데이터를 기준으로 데이터들을 같은 값들끼리 그룹화 하는 함수
        - 그룹화한 데이터에서 평균이나 합계, 최대, 최소, 표준편차, 개수와 같은 그룹화 연산을 이용할 수 있다.
    - pivot_table()
        - 데이터의 재구조화 함수
        - 데이터프레임에서 인덱스, 컬럼, 벨류, 연산식까지 원하는 방법으로 테ㅣ블 재구조화
    - 데이터프레임 결합
        - 유니언 결합(단순 결합)
            - concat()
                - 첫번째 인자에 결합할 데이터프레임들을 1차원 데이터로 입력
                    - 데이터프레임의 개수의 제한은 크게 존재하지 않음.
                - axis : 행으로 결합할 것인가 (axis=0) 열로 결합할 것인가 (axis=1)
                - pandas 안에 내장 함수, pd.concat([df1, df2])
            - 조인 결합 ( 특정 조건에 맞게 열 결합)
                - merge()
                    - 특정한 조건(특정 컬럼들끼리 같은 데이터) 기준으로 열을 결합
                    - on 매개변수가 특정한 조건인 컬럼의 이름을 지정
                    - how 매개변수는 조인의 형태를 지정
                        - lift, right, inner,outer
                    - pandas에도 DataFrame에도 해당 함수 존재. 다른 점은 첫 번째 인자가 데이터냐 아니냐.
                    - 2개의 데이터프레임을 결합하는 함수, 3개 이상의 데이터프레임을 조인결합 할때는 merge()를 여러번 사용
